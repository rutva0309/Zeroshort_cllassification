```python
#zero-shot-classification using fb model
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
#passing sentences and intents for fb model
sequence_to_classify = ["Please show all the transactions of my wallet.", "I would like to withdrawal 101 rand using EFT from my African Bank account.",
                        "Please withdrawl 101 rand from cash at pick at pay option.", "Please withdrawl 101 rand from cash at pick at pay option and send it to this number 27740779895.",
                        " Please show my account profile", " I want to Pay by Qr.", " I would like to send 10 Rand on this number 27614040684.", "Can you please buy me an airtime of 10 rand for this number 27614040684.",
                        "Please send me the list of all the data bundles.", "Please send me the list of all the electricity vouchers.", "I would like to pay from Qr code.", "I would like to pay 12 rand from my wallet.",
                        "I would like to pay 150 rand from my Bob card.", " I would like to top up of 24 rand using my virtual card.", "I would like to send payment resquest of 22 rands for Last month grocery payment on this number 27740779895."]
candidate_labels = ['transactions','withdrawl','account profile','buy airtime', 'buy electricity','pay', 'payment request']
ls =  classifier(sequence_to_classify, candidate_labels)
```

```python
#Creating DataFrame for sentences
import pandas as pd
df = pd.DataFrame(sequence_to_classify)        
df.rename(columns={0:'Sentences'}, inplace = True)
df
```

```python
#Getting labels and its scores according to the sentences
label_score = list(map(lambda x: dict(zip(x["labels"], x["scores"])),ls))
dls = pd.DataFrame(label_score)
score_fb = pd.DataFrame(dls.max(axis=1))
score_fb.rename(columns={0:'Max_Score_fb'}, inplace = True)
score_fb
```

```python
#Creating DataFrame for ntent having highest score in fb model
dfmaxfb = pd.DataFrame(dls.idxmax(axis = 1))
dfmaxfb.rename(columns={0:'Max_fb'}, inplace = True)
dfmaxfb
```

```python
#zero-shot-classification using roberta
from transformers import pipeline
clf = pipeline("zero-shot-classification",model="vicgalle/xlm-roberta-large-xnli-anli")
```

```python
#passing sentences and intents for roberta model
sent_to_classify = ["Please show all the transactions of my wallet.", "I would like to withdrawal 101 rand using EFT from my African Bank account.",
                        "Please withdrawl 101 rand from cash at pick at pay option.", "Please withdrawl 101 rand from cash at pick at pay option and send it to this number 27740779895.",
                        " Please show my account profile", " I want to Pay by Qr.", " I would like to send 10 Rand on this number 27614040684.", "Can you please buy me an airtime of 10 rand for this number 27614040684.",
                        "Please send me the list of all the data bundles.", "Please send me the list of all the electricity vouchers.", "I would like to pay from Qr code.", "I would like to pay 12 rand from my wallet.",
                        "I would like to pay 150 rand from my Bob card.", " I would like to top up of 24 rand using my virtual card.", "I would like to send payment resquest of 22 rands for Last month grocery payment on this number 27740779895."]                
candidate_labels = ['transactions','withdrawl','account profile','buy airtime', 'buy electricity','pay', 'payment request']
l =  clf(sent_to_classify, candidate_labels)
```

```python
#Getting labels and its scores according to the sentences
labels_scores = list(map(lambda x: dict(zip(x["labels"], x["scores"])),l))
di = pd.DataFrame(labels_scores)
score_r = pd.DataFrame(di.max(axis=1))
score_r.rename(columns={0:'Max_Score_Roberta'}, inplace = True)
score_r
```

```python
#Creating DataFrame for ntent having highest score in Roberta model
dfmaxr = pd.DataFrame(di.idxmax(axis = 1))
dfmaxr.rename(columns={0:'Max_Roberta'}, inplace = True)
dfmaxr
```

```python
data = pd.concat([df, dfmaxfb, score_fb, dfmaxr, score_r], axis=1)
data.to_csv("ukeshe_max_intents.csv")
data
```

```python
#Giving input for the text and forming the dictionary for the labels and scores for fb model
def get_input_fb():

    """
    returns labels and scores of the text
    
    """

    text = input("Enter: ")
    labels = ['transactions','withdrawl','account profile','buy airtime', 'buy electricity','pay', 'payment request']
    l_s = classifier(text, labels)
    return dict(zip(l_s['labels'], l_s['scores']))
```
